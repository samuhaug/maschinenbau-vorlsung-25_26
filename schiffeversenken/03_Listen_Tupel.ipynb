## Block 1: Listen (45 min)

### Listen erstellen

Eine Liste speichert mehrere Werte in einer festen Reihenfolge. Listen werden mit eckigen Klammern `[]` erstellt:

````python
# Liste erstellen
farben = ["rot", "grün", "blau"]
print(farben)

# Leere Liste
leere_liste = []
print(leere_liste)
````

### Indizierung — auf Elemente zugreifen

Jedes Element hat eine Position (Index), die bei **0** beginnt:

```
farben = ["rot", "grün", "blau"]
Index:      0       1       2
Negativ:   -3      -2      -1
```

- **Positiver Index:** Von vorne gezählt, startet bei 0
- **Negativer Index:** Von hinten gezählt, `-1` ist das letzte Element

````python
farben = ["rot", "grün", "blau"]

print(farben[0])   # Erstes Element
print(farben[2])   # Drittes Element
print(farben[-1])  # Letztes Element
````

### Slicing — Teillisten extrahieren

Mit `liste[start:stop]` kannst du einen Ausschnitt der Liste holen. Der `stop`-Index ist **exklusive** (genau wie bei `range()`).

````python
zahlen = [10, 20, 30, 40, 50]

print(zahlen[1:3])   # Element 1 und 2 → [20, 30]
print(zahlen[:3])    # Von Anfang bis Element 2 → [10, 20, 30]
print(zahlen[2:])    # Ab Element 2 bis Ende → [30, 40, 50]
````

### Veränderbarkeit (mutable)

Listen sind **veränderbar** — man kann einzelne Elemente per Index überschreiben:

````python
# Element per Index ändern
zeile = ["~", "~", "~", "~", "~"]
print(f"Vorher:  {zeile}")

zeile[2] = "X"  # Drittes Element ändern
print(f"Nachher: {zeile}")
````

### Verschachtelte Listen — 2D-Datenstruktur

Eine Liste kann andere Listen enthalten. So entsteht eine **2D-Struktur** — perfekt für ein Spielfeld:

```
spielfeld = [
    ["~", "~", "~"],     ← Zeile 0
    ["~", "~", "~"],     ← Zeile 1
    ["~", "~", "~"]      ← Zeile 2
]
```

Zugriff: `spielfeld[zeile][spalte]`

````python
# 3×3 Spielfeld als verschachtelte Liste
spielfeld = [
    ["~", "~", "~"],
    ["~", "~", "~"],
    ["~", "~", "~"]
]

# Zugriff: spielfeld[zeile][spalte]
print(f"Zeile 0, Spalte 1: {spielfeld[0][1]}")

# Wert ändern: Treffer in Zeile 1, Spalte 2 markieren
spielfeld[1][2] = "X"
print(f"Nach Treffer: {spielfeld[1]}")
````

````python
# Verschachtelte Liste mit Schleife ausgeben
spielfeld = [
    ["~", "~", "~"],
    ["~", "X", "~"],
    ["~", "~", "O"]
]

for zeile in spielfeld:
    for feld in zeile:
        print(feld, end=" ")
    print()
````

## Block 2: Methoden & Tupel (45 min)

### Die 4 wichtigsten Listen-Operationen

| Operation | Beschreibung | Beispiel |
|-----------|-------------|----------|
| `append(wert)` | Fügt ein Element am Ende hinzu | `liste.append("neu")` |
| `remove(wert)` | Entfernt das erste Vorkommen | `liste.remove("alt")` |
| `len(liste)` | Gibt die Anzahl der Elemente zurück | `len(liste)` → `5` |
| `wert in liste` | Prüft ob ein Wert enthalten ist | `"X" in liste` → `True` |

````python
# append() — Element hinzufügen
schuesse = []
schuesse.append("A1")
schuesse.append("B3")
schuesse.append("C2")
print(f"Schüsse: {schuesse}")
print(f"Anzahl: {len(schuesse)}")
````

````python
# in — Prüfen ob Element enthalten
schuesse = ["A1", "B3", "C2"]

if "B3" in schuesse:
    print("B3 wurde schon geschossen")

if "D4" not in schuesse:
    print("D4 wurde noch nicht geschossen")
````

````python
# remove() — Element entfernen
positionen = ["A1", "A2", "A3"]
print(f"Vorher: {positionen}")

positionen.remove("A2")
print(f"Nachher: {positionen}")
````

### Tupel — unveränderliche Wertepaare

Ein Tupel ist wie eine Liste, aber **unveränderlich** (immutable). Einmal erstellt, können die Werte nicht geändert werden.

Tupel werden mit runden Klammern `()` erstellt:

````python
# Tupel erstellen
koordinate = (2, 3)
print(f"Zeile: {koordinate[0]}, Spalte: {koordinate[1]}")

# Tupel entpacken
zeile, spalte = koordinate
print(f"Zeile: {zeile}, Spalte: {spalte}")
````

````python
# Tupel sind unveränderlich
koordinate = (2, 3)

# Das würde einen Fehler erzeugen:
# koordinate[0] = 5  # TypeError: 'tuple' object does not support item assignment
````

### Faustregel: Liste vs. Tupel

| | Liste | Tupel |
|---|-------|------|
| **Syntax** | `[1, 2, 3]` | `(1, 2, 3)` |
| **Veränderbar?** | Ja | Nein |
| **Einsatz** | Sammlungen die wachsen/schrumpfen | Feste Wertepaare |
| **Beispiel** | Schussliste, Spielfeldzeile | Koordinate `(2, 3)` |

**Faustregel:** Tupel für Dinge die zusammengehören und sich nicht ändern (z.B. eine Koordinate), Liste für Sammlungen die wachsen oder schrumpfen (z.B. die Liste aller Schüsse).

````python
# Praxisbeispiel: Liste von Tupeln
schuesse = []

schuesse.append((0, 1))  # Schuss auf Zeile 0, Spalte 1
schuesse.append((2, 3))  # Schuss auf Zeile 2, Spalte 3
schuesse.append((1, 4))  # Schuss auf Zeile 1, Spalte 4

print(f"Alle Schüsse: {schuesse}")

# Prüfen ob eine Koordinate schon geschossen wurde
if (2, 3) in schuesse:
    print("Zeile 2, Spalte 3 wurde schon geschossen")
````
