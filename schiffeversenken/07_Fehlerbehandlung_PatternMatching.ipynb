## Block 1: `try`/`except`/`finally` (45 min)

### Warum Fehlerbehandlung?

Ohne Fehlerbehandlung stürzt das Programm ab, wenn etwas Unerwartetes passiert — z.B. wenn ein Nutzer Text statt einer Zahl eingibt:

```python
zahl = int(input("Zahl: "))  # Eingabe: "abc" → ValueError!
```

Mit `try`/`except` können wir solche Fehler **abfangen** und sinnvoll darauf reagieren, statt das Programm abstürzen zu lassen.

### Aufbau: `try`/`except`

```python
try:
    # "Riskanter" Code, der einen Fehler erzeugen könnte
except FehlertypName:
    # Was passieren soll, wenn der Fehler auftritt
```

**Wichtig:** Immer den konkreten Fehlertyp angeben — niemals nur `except:` (das fängt alles ab, auch Programmierfehler).

### Die 3 wichtigsten Exceptions

| Exception | Wann? | Beispiel |
|-----------|-------|----------|
| `ValueError` | Falscher Wert (z.B. Text statt Zahl) | `int("abc")` |
| `FileNotFoundError` | Datei existiert nicht | `open("gibts_nicht.txt", "r")` |
| `KeyError` | Key existiert nicht im Dictionary | `dict["gibts_nicht"]` |

````python
# ValueError abfangen
eingabe = "abc"  # Simuliert eine falsche Eingabe

try:
    zahl = int(eingabe)
    print(f"Eingegebene Zahl: {zahl}")
except ValueError:
    print(f"Fehler: '{eingabe}' ist keine gültige Zahl!")
````

### `finally` — Aufräumen

`finally` wird **immer** ausgeführt — egal ob ein Fehler auftrat oder nicht. Nützlich für Aufräumarbeiten:

````python
# finally wird immer ausgeführt
try:
    zahl = int("42")
    print(f"Zahl: {zahl}")
except ValueError:
    print("Keine gültige Zahl!")
finally:
    print("Dieser Code läuft immer.")
````

### Muster: Eingabevalidierung mit `try`/`except`

Kombination mit `while True` für robuste Eingaben:

````python
# Robuste Zahleneingabe — Kombination von while + try/except
# Zum Testen: gib erst "abc" ein, dann eine Zahl

while True:
    eingabe = input("Gib eine Zahl ein: ")
    try:
        zahl = int(eingabe)
        print(f"Danke! Deine Zahl ist {zahl}.")
        break
    except ValueError:
        print(f"'{eingabe}' ist keine Zahl. Versuche es nochmal.")
````

## Block 2: Pattern Matching (45 min)

### Was ist Pattern Matching?

`match`/`case` ist eine lesbare Alternative zu langen `if`/`elif`-Ketten, wenn man eine Variable gegen **feste Werte** prüfen will.

```python
match variable:
    case "wert1":
        # Code für wert1
    case "wert2":
        # Code für wert2
    case _:
        # Default: keiner der obigen Werte
```

- `case _:` ist der **Wildcard** (Default-Fall) — greift wenn kein anderer Fall passt
- Funktioniert mit Strings, Zahlen und anderen Literalen

````python
# Einfaches match/case
farbe = "rot"

match farbe:
    case "rot":
        print("Stopp!")
    case "gelb":
        print("Achtung!")
    case "grün":
        print("Fahren!")
    case _:
        print(f"Unbekannte Farbe: {farbe}")
````

**🔧 Mini-Aufgabe:** Ergänze das `match`/`case` um einen Fall `"blau"` mit der Ausgabe `"Parken!"`. Ändere `farbe` auf `"blau"` und teste, dass der neue Fall greift.

### Einsatz: Menüsteuerung

`match`/`case` ist ideal für Menüs, in denen der Nutzer eine Option wählt:

````python
# Menü mit match/case
print("=== Hauptmenü ===")
print("1) Neues Spiel")
print("2) Spielstand laden")
print("3) Beenden")

wahl = input("Wähle eine Option: ")

match wahl:
    case "1":
        print("Neues Spiel wird gestartet...")
    case "2":
        print("Spielstand wird geladen...")
    case "3":
        print("Auf Wiedersehen!")
    case _:
        print("Ungültige Eingabe!")
````

### Vergleich: `match`/`case` vs. `if`/`elif`

| | `match`/`case` | `if`/`elif` |
|---|----------------|-------------|
| **Wann?** | Diskrete Werte (Menü, Status) | Bereiche, Bedingungen |
| **Lesbarkeit** | Sehr gut bei vielen Fällen | Kann unübersichtlich werden |
| **Beispiel** | Nutzer wählt Option 1/2/3 | Zahl liegt im Bereich 1–10 |

````python
# Vergleich: Gleiches Ergebnis, andere Schreibweise
ergebnis = "treffer"

# Mit match/case (gut lesbar bei festen Werten)
match ergebnis:
    case "treffer":
        print("🎯 Treffer!")
    case "wasser":
        print("🌊 Wasser!")
    case "versenkt":
        print("💥 Versenkt!")
    case _:
        print("Unbekannt")
````

````python
# Gleiches mit if/elif (funktioniert auch, aber bei vielen Fällen weniger übersichtlich)
ergebnis = "treffer"

if ergebnis == "treffer":
    print("🎯 Treffer!")
elif ergebnis == "wasser":
    print("🌊 Wasser!")
elif ergebnis == "versenkt":
    print("💥 Versenkt!")
else:
    print("Unbekannt")
````
