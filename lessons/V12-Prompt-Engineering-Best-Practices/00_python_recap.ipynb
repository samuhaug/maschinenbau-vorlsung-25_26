## 1. Typen prüfen mit `type()` und `isinstance()` (V03)

- `type(x)` liefert den Datentyp eines Wertes (z. B. `int`, `float`, `str`, `bool`).
- `isinstance(x, typ)` prüft, ob `x` von diesem Typ ist. Das Ergebnis ist `True` oder `False`.
- `isinstance` akzeptiert auch ein Tupel von Typen: `isinstance(x, (int, float))`.

```python
drehzahl = 1500
vorschub = 0.25
maschine = "CNC-Fraese-01"

print(type(drehzahl).__name__, type(vorschub).__name__, type(maschine).__name__)
print("drehzahl ist int?        ", isinstance(drehzahl, int))
print("vorschub ist Zahl?       ", isinstance(vorschub, (int, float)))
print("maschine ist int?        ", isinstance(maschine, int))
```

## 2. `bool()` als Cast + Truthy/Falsy (V03)

`bool(x)` wandelt einen Wert in `True` oder `False` um. Merksatz: **„leer / 0 / None" ist
falsy, alles andere truthy.**

| Falsy (= `bool(x) == False`) | Truthy (= `bool(x) == True`) |
|---|---|
| `0`, `0.0` | Jede Zahl ≠ 0 |
| `""` (leerer String) | Jeder nicht-leere String |
| `[]`, `{}`, `()` (leere Container) | Nicht-leere Container |
| `None`, `False` | `True`, Objekte |

```python
werte = [0, 1, -3, 0.0, 0.1, "", "cnc", [], [0], None]
for w in werte:
    print(f"{repr(w):>8}  ->  bool = {bool(w)}")
```

## 3. `round()`, `min()`, `max()` (V03)

- `round(x, n)` rundet `x` auf `n` Nachkommastellen (`n` fehlt → ganze Zahl).
- `min(a, b, ...)` / `max(a, b, ...)` liefern das kleinste bzw. größte Argument.
- Beide akzeptieren auch eine Liste: `min([3, 1, 4])` → `1`.

```python
messwerte = [22.7, 21.4, 23.9, 22.1, 24.3]
print("gerundet (1 Stelle):", round(sum(messwerte) / len(messwerte), 1))
print("min / max          :", min(messwerte), "/", max(messwerte))
print("round(3.14159, 2)  :", round(3.14159, 2))
```

## 4. Vergleichsoperatoren (V04)

| Operator | Bedeutung | Beispiel |
|---|---|---|
| `==` | gleich | `5 == 5.0` → `True` |
| `!=` | ungleich | `"a" != "b"` → `True` |
| `<`, `<=` | kleiner, kleiner-gleich | `3 < 5` → `True` |
| `>`, `>=` | größer, größer-gleich | `7 >= 7` → `True` |

Das Ergebnis eines Vergleichs ist **immer** `True` oder `False`.

```python
temperatur = 78.5
schwelle = 80.0
print("Temp < Schwelle :", temperatur < schwelle)
print("Temp == 78.5    :", temperatur == 78.5)
print("Temp != Schwelle:", temperatur != schwelle)
```

## 5. Logische Operatoren, verkettete Vergleiche, Kurzschluss (V04)

- `and` – **beide** Seiten müssen wahr sein.
- `or`  – **mindestens eine** Seite muss wahr sein.
- `not` – kehrt den Wahrheitswert um.
- **Verkettete Vergleiche** wie `0 < x < 10` sind in Python erlaubt und entsprechen
  `(0 < x) and (x < 10)`.
- **Kurzschlussauswertung**: `False and …` liefert sofort `False`, ohne die rechte
  Seite zu prüfen. Genauso liefert `True or …` sofort `True`. Das schützt z. B. vor
  Division durch 0: `x != 0 and 10 / x > 1`.

```python
temperatur = 78.5
druck = 4.2

im_soll_bereich = 60 <= temperatur <= 80        # verketteter Vergleich
alarm = temperatur > 80 or druck > 5.0           # or
alles_ok = im_soll_bereich and not alarm         # and + not

print("Temperatur im Soll:", im_soll_bereich)
print("Alarm-Zustand    :", alarm)
print("Alles ok          :", alles_ok)

# Kurzschluss schützt vor Division durch 0:
x = 0
sicher = x != 0 and 10 / x > 1
print("sicher (x=0)      :", sicher)
```

## ✅ Cheat-Sheet

| Werkzeug | Zweck | Mini-Beispiel |
|---|---|---|
| `type(x)` | Datentyp anschauen | `type(3) is int` |
| `isinstance(x, T)` | Typ prüfen (True/False) | `isinstance(3, int)` |
| `bool(x)` | Wert zu True/False casten | `bool("") == False` |
| Truthy/Falsy | leer/0/None = falsy | `if liste:` |

| `min` / `max` | Extremwerte | `min(3, 1, 4)` → `1` |
| `==`, `!=`, `<`, `<=`, `>`, `>=` | Vergleiche | `temp < 80` |
| `and`, `or`, `not` | Logik | `a and not b` |
| `a < x < b` | verketteter Vergleich | `0 < x < 10` |
| Kurzschluss | rechte Seite bleibt ungeprüft | `x != 0 and 10/x > 1` |
