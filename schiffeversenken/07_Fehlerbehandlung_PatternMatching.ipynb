## Block 1: `try`/`except`/`finally` (45 min)

### Warum Fehlerbehandlung?

Ohne Fehlerbehandlung stĂĽrzt das Programm ab, wenn etwas Unerwartetes passiert â€” z.B. wenn ein Nutzer Text statt einer Zahl eingibt:

```python
zahl = int(input("Zahl: "))  # Eingabe: "abc" â†’ ValueError!
```

Mit `try`/`except` kĂ¶nnen wir solche Fehler **abfangen** und sinnvoll darauf reagieren, statt das Programm abstĂĽrzen zu lassen.

### Aufbau: `try`/`except`

```python
try:
    # "Riskanter" Code, der einen Fehler erzeugen kĂ¶nnte
except FehlertypName:
    # Was passieren soll, wenn der Fehler auftritt
```

**Wichtig:** Immer den konkreten Fehlertyp angeben â€” niemals nur `except:` (das fĂ¤ngt alles ab, auch Programmierfehler).

### Die 3 wichtigsten Exceptions

| Exception | Wann? | Beispiel |
|-----------|-------|----------|
| `ValueError` | Falscher Wert (z.B. Text statt Zahl) | `int("abc")` |
| `FileNotFoundError` | Datei existiert nicht | `open("gibts_nicht.txt", "r")` |
| `KeyError` | Key existiert nicht im Dictionary | `dict["gibts_nicht"]` |

````python
# ValueError abfangen
eingabe = "abc"  # Simuliert eine falsche Eingabe

try:
    zahl = int(eingabe)
    print(f"Eingegebene Zahl: {zahl}")
except ValueError:
    print(f"Fehler: '{eingabe}' ist keine gĂĽltige Zahl!")
````

````python
# FileNotFoundError abfangen
try:
    with open("gibts_nicht.txt", "r") as f:
        inhalt = f.read()
except FileNotFoundError:
    print("Datei nicht gefunden!")
````

````python
# KeyError abfangen
spieler = {"name": "Anna", "punkte": 42}

try:
    wert = spieler["level"]
except KeyError:
    print("Key 'level' existiert nicht im Dictionary!")
````

### `finally` â€” AufrĂ¤umen

`finally` wird **immer** ausgefĂĽhrt â€” egal ob ein Fehler auftrat oder nicht. NĂĽtzlich fĂĽr AufrĂ¤umarbeiten:

````python
# finally wird immer ausgefĂĽhrt
try:
    zahl = int("42")
    print(f"Zahl: {zahl}")
except ValueError:
    print("Keine gĂĽltige Zahl!")
finally:
    print("Dieser Code lĂ¤uft immer.")
````

### Muster: Eingabevalidierung mit `try`/`except`

Kombination mit `while True` fĂĽr robuste Eingaben:

````python
# Robuste Zahleneingabe â€” Kombination von while + try/except
# Zum Testen: gib erst "abc" ein, dann eine Zahl

while True:
    eingabe = input("Gib eine Zahl ein: ")
    try:
        zahl = int(eingabe)
        print(f"Danke! Deine Zahl ist {zahl}.")
        break
    except ValueError:
        print(f"'{eingabe}' ist keine Zahl. Versuche es nochmal.")
````

## Block 2: Pattern Matching (45 min)

### Was ist Pattern Matching?

`match`/`case` ist eine lesbare Alternative zu langen `if`/`elif`-Ketten, wenn man eine Variable gegen **feste Werte** prĂĽfen will.

```python
match variable:
    case "wert1":
        # Code fĂĽr wert1
    case "wert2":
        # Code fĂĽr wert2
    case _:
        # Default: keiner der obigen Werte
```

- `case _:` ist der **Wildcard** (Default-Fall) â€” greift wenn kein anderer Fall passt
- Funktioniert mit Strings, Zahlen und anderen Literalen

````python
# Einfaches match/case
farbe = "rot"

match farbe:
    case "rot":
        print("Stopp!")
    case "gelb":
        print("Achtung!")
    case "grĂĽn":
        print("Fahren!")
    case _:
        print(f"Unbekannte Farbe: {farbe}")
````

### Einsatz: MenĂĽsteuerung

`match`/`case` ist ideal fĂĽr MenĂĽs, in denen der Nutzer eine Option wĂ¤hlt:

````python
# MenĂĽ mit match/case
print("=== HauptmenĂĽ ===")
print("1) Neues Spiel")
print("2) Spielstand laden")
print("3) Beenden")

wahl = input("WĂ¤hle eine Option: ")

match wahl:
    case "1":
        print("Neues Spiel wird gestartet...")
    case "2":
        print("Spielstand wird geladen...")
    case "3":
        print("Auf Wiedersehen!")
    case _:
        print("UngĂĽltige Eingabe!")
````

### Vergleich: `match`/`case` vs. `if`/`elif`

| | `match`/`case` | `if`/`elif` |
|---|----------------|-------------|
| **Wann?** | Diskrete Werte (MenĂĽ, Status) | Bereiche, Bedingungen |
| **Lesbarkeit** | Sehr gut bei vielen FĂ¤llen | Kann unĂĽbersichtlich werden |
| **Beispiel** | Nutzer wĂ¤hlt Option 1/2/3 | Zahl liegt im Bereich 1â€“10 |

````python
# Vergleich: Gleiches Ergebnis, andere Schreibweise
ergebnis = "treffer"

# Mit match/case (gut lesbar bei festen Werten)
match ergebnis:
    case "treffer":
        print("đźŽŻ Treffer!")
    case "wasser":
        print("đźŚŠ Wasser!")
    case "versenkt":
        print("đź’Ą Versenkt!")
    case _:
        print("Unbekannt")
````

````python
# Gleiches mit if/elif (funktioniert auch, aber bei vielen FĂ¤llen weniger ĂĽbersichtlich)
ergebnis = "treffer"

if ergebnis == "treffer":
    print("đźŽŻ Treffer!")
elif ergebnis == "wasser":
    print("đźŚŠ Wasser!")
elif ergebnis == "versenkt":
    print("đź’Ą Versenkt!")
else:
    print("Unbekannt")
````
