## Block 1: Imports (45 min)

### Module importieren

Python bringt viele nützliche Module mit, die zusätzliche Funktionen bereitstellen. Es gibt zwei Import-Formen:

| Form | Beispiel | Aufruf |
|------|----------|--------|
| `import modul` | `import random` | `random.randint(0, 4)` |
| `from modul import name` | `from random import randint` | `randint(0, 4)` |

**Empfehlung:** `import modul` ist meistens besser, weil man sofort sieht woher eine Funktion kommt.

````python
# import modul — Modul als Ganzes importieren
import random

# Zufällige Zahl zwischen 0 und 4 (inklusive)
zufallszahl = random.randint(0, 4)
print(f"Zufallszahl: {zufallszahl}")
````

````python
# from modul import name — einzelne Funktion importieren
from random import randint

zufallszahl = randint(0, 4)
print(f"Zufallszahl: {zufallszahl}")
````

### Eigene Module

## Block 2: Dateien lesen & schreiben (45 min)

### `with open()` — der einzige Weg

Dateien werden in Python immer mit `with open()` geöffnet. Das `with` sorgt dafür, dass die Datei automatisch geschlossen wird:

```python
with open("datei.txt", modus) as f:
    # Hier mit der Datei arbeiten
# Datei wird automatisch geschlossen
```

**Modi:**

| Modus | Bedeutung | Verhalten |
|-------|-----------|----------|
| `"r"` | Lesen (read) | Datei muss existieren |
| `"w"` | Schreiben (write) | Erstellt neue Datei / überschreibt bestehende |

### 📝 Mini-Aufgabe

Öffne `test.txt` im Lesemodus mit `with open()` und gib den kompletten Inhalt mit `f.read()` aus.

````python
with open("test.txt", "r") as f:
    inhalt = f.read()
    print(inhalt)
````

### In eine Datei schreiben

````python
# In eine Datei schreiben
with open("test.txt", "w") as f:
    f.write("Anna\n")
    f.write("Max\n")
    f.write("Lisa\n")

print("Datei geschrieben!")
````

### Aus einer Datei lesen

Zeilenweises Lesen mit einer `for`-Schleife. **Wichtig:** Jede Zeile enthält am Ende ein `\n` (Zeilenumbruch), das mit `.strip()` entfernt werden sollte.

````python
# Datei zeilenweise lesen
with open("test.txt", "r") as f:
    for zeile in f:
        name = zeile.strip()  # Zeilenumbruch entfernen
        print(f"Gelesen: {name}")
````

### CSV-Dateien

CSV (Comma-Separated Values) ist ein einfaches Format für tabellarische Daten. Python hat dafür das `csv`-Modul:

````python
import csv

# CSV schreiben
with open("noten.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Name", "Note"])
    writer.writerow(["Anna", 1])
    writer.writerow(["Max", 2])
    writer.writerow(["Lisa", 1])

print("CSV geschrieben!")
````

````python
import csv

# CSV lesen
with open("noten.csv", "r") as f:
    reader = csv.reader(f)
    for zeile in reader:
        print(zeile)  # Jede Zeile ist eine Liste von Strings
````

### Aufgabe 3: `spielstand_laden(dateiname)`

Schreibe eine Funktion, die die CSV-Datei einliest und Spielfeld + Statistik zurückgibt.

- Zeile 1 → Statistik-Dictionary rekonstruieren
- Zeilen 2–6 → Spielfeld als 2D-Liste rekonstruieren
- Rückgabe: `spielfeld, statistik` (zwei Werte)

Teste: Speichern → Laden → Spielfeld ausgeben.

<details>
<summary>💡 Tipp</summary>

Beim Lesen mit `csv.reader()` sind alle Werte Strings. Die Statistik-Werte müssen mit `int()` umgewandelt werden.

Zwei Werte zurückgeben: `return spielfeld, statistik`

Zwei Werte empfangen: `spielfeld, statistik = spielstand_laden("spielstand.csv")`
</details>

````python
import csv
from spiellogik import spielfeld_ausgeben

def spielstand_laden(dateiname):
    spielfeld = []
    with open(dateiname, "r") as f:
        reader = csv.reader(f)
        # Erste Zeile: Statistik
        erste_zeile = next(reader)
        statistik = {
            "treffer": int(erste_zeile[0]),
            "wasser": int(erste_zeile[1]),
            "versenkt": int(erste_zeile[2])
        }
        # Restliche Zeilen: Spielfeld
        for zeile in reader:
            spielfeld.append(zeile)
    return spielfeld, statistik

# Test: Laden und ausgeben
geladenes_spielfeld, geladene_statistik = spielstand_laden("spielstand.csv")
print("Geladenes Spielfeld:")
spielfeld_ausgeben(geladenes_spielfeld)
print(f"Statistik: {geladene_statistik}")
````

### Aufgabe 4: Zufällige Schiffsplatzierung

Schreibe eine Funktion `platziere_schiffe()`, die Schiffe zufällig auf dem 5×5-Feld platziert:

1. Nutze `random.randint(0, 4)` für die Startposition
2. Wähle eine zufällige Richtung: horizontal (Spalte erhöht sich) oder vertikal (Zeile erhöht sich)
3. Prüfe ob das Schiff ins Feld passt (nicht über den Rand hinaus)
4. Platziere zwei Schiffe: Kreuzer (3 Felder) und U-Boot (2 Felder)
5. Rückgabe: Dictionary wie in VL 4

<details>
<summary>💡 Tipp</summary>

Richtung zufällig wählen:
```python
richtung = random.randint(0, 1)  # 0 = horizontal, 1 = vertikal
```

Passt das Schiff? Bei horizontal: `spalte + laenge - 1 <= 4`, bei vertikal: `zeile + laenge - 1 <= 4`.

Falls es nicht passt: Neue Zufallswerte ziehen (mit `while`-Schleife).
</details>

````python
import random

def platziere_schiffe():
    schiffe = {}
    schiffstypen = [("kreuzer", 3), ("u_boot", 2)]

    for name, laenge in schiffstypen:
        platziert = False
        while not platziert:
            zeile = random.randint(0, 4)
            spalte = random.randint(0, 4)
            richtung = random.randint(0, 1)  # 0 = horizontal, 1 = vertikal

            if richtung == 0 and spalte + laenge - 1 <= 4:
                # Horizontal: Spalte erhöht sich
                koordinaten = [(zeile, spalte + i) for i in range(laenge)]
                platziert = True
            elif richtung == 1 and zeile + laenge - 1 <= 4:
                # Vertikal: Zeile erhöht sich
                koordinaten = [(zeile + i, spalte) for i in range(laenge)]
                platziert = True

        schiffe[name] = {"koordinaten": koordinaten, "versenkt": False}

    return schiffe

# Test
schiffe = platziere_schiffe()
print("Schiffe platziert:")
for name, daten in schiffe.items():
    print(f"  {name}: {daten['koordinaten']}")
````

### Aufgabe 5: Integration

Führe alles zusammen:

1. Starte ein neues Spiel mit zufällig platzierten Schiffen (`platziere_schiffe()`)
2. Erstelle ein Spielfeld und spiele ein paar Züge (simuliert mit festen Koordinaten)
3. Speichere den Spielstand mit `spielstand_speichern()`
4. Lade den Spielstand in einer neuen Zelle mit `spielstand_laden()`
5. Gib das geladene Spielfeld aus — es muss identisch mit dem gespeicherten sein

<details>
<summary>💡 Tipp</summary>

```python
# Neues Spiel
spielfeld = erstelle_spielfeld()
schiffe = platziere_schiffe()
statistik = {"treffer": 0, "wasser": 0, "versenkt": 0}

# Ein paar Züge simulieren...
# ...

# Speichern
spielstand_speichern("spielstand.csv", spielfeld, statistik)

# Laden und vergleichen
geladenes_spielfeld, geladene_statistik = spielstand_laden("spielstand.csv")
spielfeld_ausgeben(geladenes_spielfeld)
```
</details>

````python
from spiellogik import erstelle_spielfeld, spielfeld_ausgeben

# 1. Neues Spiel starten
spielfeld = erstelle_spielfeld()
schiffe = platziere_schiffe()
statistik = {"treffer": 0, "wasser": 0, "versenkt": 0}

print("Schiffe platziert:")
for name, daten in schiffe.items():
    print(f"  {name}: {daten['koordinaten']}")

# 2. Züge simulieren: erste Koordinate des Kreuzers treffen, dann Wasser
treffer_koordinate = schiffe["kreuzer"]["koordinaten"][0]
zeile_t, spalte_t = treffer_koordinate
spielfeld[zeile_t][spalte_t] = "X"
statistik["treffer"] += 1

# Wasser: Koordinate (4, 4) schießen (nur wenn kein Schiff dort)
spielfeld[4][4] = "O"
statistik["wasser"] += 1

print("\nSpielfeld nach Zügen:")
spielfeld_ausgeben(spielfeld)
print(f"Statistik: {statistik}")

# 3. Spielstand speichern
spielstand_speichern("spielstand.csv", spielfeld, statistik)
print("\nSpielstand gespeichert.")

# 4. Spielstand laden
geladenes_spielfeld, geladene_statistik = spielstand_laden("spielstand.csv")

# 5. Vergleich
print("\nGeladenes Spielfeld:")
spielfeld_ausgeben(geladenes_spielfeld)
print(f"Geladene Statistik: {geladene_statistik}")
````
