## Block 1: Wiederholung `if`/`elif`/`else` (45 min)

### Grundsyntax

Eine Verzweigung prüft eine Bedingung und führt je nach Ergebnis unterschiedlichen Code aus.

```python
if bedingung:
    # wird ausgeführt wenn bedingung True ist
elif andere_bedingung:
    # wird ausgeführt wenn andere_bedingung True ist
else:
    # wird ausgeführt wenn keine Bedingung True war
```

**Wichtige Regeln:**
- Nach `if`, `elif`, `else` steht immer ein **Doppelpunkt** `:`
- Der zugehörige Code-Block muss **eingerückt** sein (4 Leerzeichen)
- `elif` und `else` sind optional
- Es können beliebig viele `elif`-Zweige folgen, aber nur ein `else`

````python
# Beispiel: Zahl klassifizieren
zahl = 7

if zahl > 0:
    print(f"{zahl} ist positiv")
elif zahl < 0:
    print(f"{zahl} ist negativ")
else:
    print("Die Zahl ist null")
````

### Häufiger Fehler: `=` vs. `==`

| Operator | Bedeutung | Beispiel |
|----------|-----------|----------|
| `=` | **Zuweisung** — speichert einen Wert | `x = 5` |
| `==` | **Vergleich** — prüft auf Gleichheit | `x == 5` → `True` |

Ein `=` in einer `if`-Bedingung ist ein **Syntaxfehler** in Python:

````python
# Richtig: Vergleich mit ==
spieler = "Anna"

if spieler == "Anna":
    print("Hallo Anna!")
````

### Wann `elif` statt verschachteltem `if`?

Stell dir eine Verzweigung wie einen **Entscheidungsbaum** vor: Von oben nach unten wird jede Bedingung geprüft, bis eine zutrifft.

**Faustregel:** Wenn die Bedingungen sich gegenseitig ausschließen (nur eine kann gleichzeitig wahr sein), verwende `elif`. Verschachtelte `if`-Blöcke machen den Code schwerer lesbar.

````python
# Gut: elif für sich ausschließende Bedingungen
note = 2

if note == 1:
    print("Sehr gut")
elif note == 2:
    print("Gut")
elif note == 3:
    print("Befriedigend")
elif note == 4:
    print("Ausreichend")
else:
    print("Nicht bestanden")
````

````python
# Schlecht: Verschachtelte if-Blöcke für das gleiche Problem
note = 2

if note == 1:
    print("Sehr gut")
else:
    if note == 2:
        print("Gut")
    else:
        if note == 3:
            print("Befriedigend")
        else:
            print("Nicht bestanden")

# → Schwerer zu lesen, gleiche Logik wie oben
````

### Truthy und Falsy

Python wertet nicht nur `True` und `False` als Wahrheitswerte aus. Bestimmte Werte gelten als **falsy** (werden wie `False` behandelt):

| Falsy-Werte | Beispiel |
|-------------|----------|
| `False` | `if False:` |
| `0` | `if 0:` |
| `""` (leerer String) | `if "":` |
| `None` | `if None:` |

Alles andere ist **truthy** (wird wie `True` behandelt): z.B. `1`, `"Hallo"`, `42`.

````python
# Truthy/Falsy in Aktion
eingabe = ""

if eingabe:
    print(f"Eingabe war: {eingabe}")
else:
    print("Keine Eingabe gemacht (leerer String ist falsy)")
````

````python
# Praktisch: Prüfen ob eine Variable einen Wert hat
name = "Max"

if name:
    print(f"Willkommen, {name}!")
else:
    print("Kein Name angegeben")
````

## Block 2: Logische Verknüpfungen (45 min)

### `and`, `or`, `not`

Mit logischen Operatoren können mehrere Bedingungen kombiniert werden:

| Operator | Bedeutung | Ergebnis `True` wenn... |
|----------|-----------|------------------------|
| `and` | UND | **beide** Bedingungen `True` sind |
| `or` | ODER | **mindestens eine** Bedingung `True` ist |
| `not` | NICHT | die Bedingung `False` ist |

**Auswertungsreihenfolge:** `not` → `and` → `or`

Im Zweifel: **Klammern setzen**, um die Reihenfolge klar zu machen.

````python
# and: Beide Bedingungen müssen True sein
alter = 25
hat_fuehrerschein = True

if alter >= 18 and hat_fuehrerschein:
    print("Darf Auto fahren")
else:
    print("Darf nicht Auto fahren")
````

````python
# or: Mindestens eine Bedingung muss True sein
ist_student = True
ist_rentner = False

if ist_student or ist_rentner:
    print("Ermäßigung möglich")
else:
    print("Normaler Preis")
````

````python
# not: Bedingung umkehren
ist_gesperrt = False

if not ist_gesperrt:
    print("Zugang erlaubt")
````

### Short-Circuit-Evaluation

Python wertet logische Ausdrücke **von links nach rechts** aus und bricht ab, sobald das Ergebnis feststeht:

- `and`: Wenn die erste Bedingung `False` ist, wird die zweite gar nicht geprüft (Ergebnis ist sicher `False`)
- `or`: Wenn die erste Bedingung `True` ist, wird die zweite gar nicht geprüft (Ergebnis ist sicher `True`)

Das ist besonders nützlich, um Fehler zu vermeiden:

````python
# Short-Circuit: Zweite Bedingung wird nur geprüft wenn nötig
text = ""

# Erst prüfen ob text nicht leer ist, DANN auf erstes Zeichen zugreifen
if text and text[0] == "A":
    print("Text beginnt mit A")
else:
    print("Text ist leer oder beginnt nicht mit A")

# Ohne Short-Circuit würde text[0] bei leerem String einen Fehler erzeugen!
````

### Komplexe Bedingungen lesbar formulieren

**Tipp 1:** Sprechende Variablen für Teilbedingungen verwenden

````python
# Schlecht lesbar:
spalte = "C"
zeile = 3

if spalte >= "A" and spalte <= "E" and zeile >= 1 and zeile <= 5:
    print("Gültige Koordinate")
````

````python
# Besser lesbar mit sprechenden Variablen:
spalte = "C"
zeile = 3

ist_gueltige_spalte = spalte >= "A" and spalte <= "E"
ist_gueltige_zeile = zeile >= 1 and zeile <= 5

if ist_gueltige_spalte and ist_gueltige_zeile:
    print("Gültige Koordinate")
else:
    print("Ungültige Koordinate")
````

### Vergleichsketten

Python erlaubt eine elegante Kurzschreibweise für Bereichsprüfungen:

````python
# Vergleichskette: kürzer und lesbarer
x = 7

# Statt: x >= 1 and x <= 10
if 1 <= x <= 10:
    print(f"{x} liegt im Bereich 1 bis 10")
````

### Typische Muster

| Muster | Beispiel | Einsatz |
|--------|----------|--------|
| **Bereichsprüfung** | `1 <= x <= 10` | Ist ein Wert im gültigen Bereich? |
| **Mehrfachbedingung** | `a and b and c` | Alle Kriterien müssen erfüllt sein |
| **Ausschlusslogik** | `not ist_gesperrt` | Etwas darf NICHT der Fall sein |

````python
# Schiffe-versenken-Kontext: Koordinatenvalidierung
spalte = "B"
zeile = 3

ist_gueltig = "A" <= spalte <= "E" and 1 <= zeile <= 5

if ist_gueltig:
    print(f"Koordinate {spalte}{zeile} ist gültig")
else:
    print(f"Koordinate {spalte}{zeile} ist ungültig")
````

### Aufgabe 5: Zusammenführen

Kombiniere alles zu einem vollständigen Programm:

1. Koordinate einlesen
2. Eingabe validieren
3. Bei gültiger Eingabe: Treffer oder Wasser prüfen
4. Passende Meldung ausgeben

Teste mit verschiedenen Eingaben: `"B2"` (Treffer), `"A1"` (Wasser), `"Z9"` (ungültig).

<details>
<summary>💡 Tipp</summary>

Verschachtele die Prüfungen: Erst Validierung im äußeren `if`, dann Trefferprüfung im inneren `if`.
</details>

````python

# ── Schiffspositionen festlegen ──────────────────────────────────────────────
schiff_1 = "B2"
schiff_2 = "D4"

# ── Schritt 1: Koordinate einlesen ───────────────────────────────────────────
eingabe = "B3"

buchstabe = eingabe[0]       # erstes Zeichen  → Spalte,  z.B. "B"
zahl      = int(eingabe[1])  # zweites Zeichen → Zeile,   z.B. 2

# ── Schritt 2: Eingabe validieren ────────────────────────────────────────────
ist_gueltige_spalte = "A" <= buchstabe <= "E"
ist_gueltige_zeile  = 1 <= zahl <= 5

if ist_gueltige_spalte and ist_gueltige_zeile:

    # ── Schritt 3: Treffer oder Wasser? ──────────────────────────────────────
    if eingabe == schiff_1 or eingabe == schiff_2:
        print("Treffer!")
    else:
        print("Wasser!")

else:
    print("Ungültige Eingabe!")
````
