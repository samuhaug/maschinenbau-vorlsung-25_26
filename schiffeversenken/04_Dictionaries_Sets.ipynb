## Block 1: Dictionaries (45 min)

### Was ist ein Dictionary?

Ein Dictionary speichert **Key-Value-Paare** — jeder Schlüssel (Key) ist mit einem Wert (Value) verknüpft. Statt über einen Index (wie bei Listen) greifst du über einen **Namen** zu.

Dictionaries werden mit geschweiften Klammern `{}` erstellt:

````python
# Dictionary erstellen
steckbrief = {
    "name": "Anna",
    "alter": 22,
    "studiengang": "Informatik"
}

print(steckbrief)
````

### Zugriff per Key

Mit `dict[key]` greifst du auf den Wert eines Schlüssels zu:

````python
steckbrief = {"name": "Anna", "alter": 22, "studiengang": "Informatik"}

# Wert lesen
print(f"Name: {steckbrief['name']}")
print(f"Alter: {steckbrief['alter']}")
````

### Hinzufügen und Ändern

Neue Einträge werden einfach per Zuweisung hinzugefügt. Existiert der Key schon, wird der Wert überschrieben:

````python
steckbrief = {"name": "Anna", "alter": 22}

# Neuen Key hinzufügen
steckbrief["stadt"] = "München"
print(f"Nach Hinzufügen: {steckbrief}")

# Bestehenden Wert ändern
steckbrief["alter"] = 23
print(f"Nach Ändern: {steckbrief}")
````

### Iteration über Dictionaries

Es gibt zwei Wege, über ein Dictionary zu iterieren:

- `for key in dict:` — nur über die Schlüssel
- `for key, value in dict.items():` — über Schlüssel **und** Werte

````python
steckbrief = {"name": "Anna", "alter": 22, "stadt": "München"}

# Über Keys iterieren
for key in steckbrief:
    print(key)
````

````python
# Über Keys und Values iterieren mit .items()
steckbrief = {"name": "Anna", "alter": 22, "stadt": "München"}

for key, value in steckbrief.items():
    print(f"{key}: {value}")
````

### Prüfen ob ein Key existiert

Mit `in` prüfst du, ob ein Schlüssel im Dictionary vorhanden ist:

````python
steckbrief = {"name": "Anna", "alter": 22}

if "name" in steckbrief:
    print("Key 'name' existiert")

if "stadt" not in steckbrief:
    print("Key 'stadt' existiert noch nicht")
````

### Faustregel: Dictionary vs. Liste

- **Liste:** Zugriff per **Position** (Index) — z.B. Spielfeldzeile, Reihenfolge wichtig
- **Dictionary:** Zugriff per **Name** (Key) — z.B. Schiffe nach Namen ansprechen

````python
# Schiffe-versenken-Kontext: Schiffe als Dictionary
schiffe = {
    "Kreuzer": [(0, 1), (0, 2), (0, 3)],
    "U-Boot": [(3, 0), (4, 0)]
}

# Zugriff per Name
print(f"Positionen des Kreuzers: {schiffe['Kreuzer']}")
print(f"Positionen des U-Boots: {schiffe['U-Boot']}")
````

## Block 2: Sets & Datenstruktur-Überblick (45 min)

### Was ist ein Set?

Ein Set ist eine **ungeordnete Sammlung einzigartiger Elemente**. Duplikate werden automatisch entfernt.

Sets werden mit `set()` erstellt (oder mit `{}` wenn Elemente vorhanden sind — aber `{}` ohne Elemente erzeugt ein Dictionary!):

````python
# Set erstellen — Duplikate werden entfernt
staedte = {"Berlin", "München", "Berlin", "Hamburg", "München"}
print(f"Set: {staedte}")
print(f"Anzahl: {len(staedte)}")
````

````python
# Leeres Set erstellen — NICHT mit {} (das wäre ein Dict)
besucht = set()
print(f"Leeres Set: {besucht}")
print(f"Typ: {type(besucht)}")
````

### Die 2 wichtigsten Set-Operationen

| Operation | Beschreibung | Beispiel |
|-----------|-------------|----------|
| `add(wert)` | Fügt ein Element hinzu (ignoriert Duplikate) | `mein_set.add("neu")` |
| `wert in mein_set` | Prüft ob ein Wert enthalten ist | `"X" in mein_set` → `True` |

**Einsatz:** Immer wenn du die Frage stellen willst: *"Wurde das schon gemacht?"*

````python
# add() und in
besucht = set()

besucht.add("Berlin")
besucht.add("München")
besucht.add("Berlin")  # Duplikat wird ignoriert

print(f"Besucht: {besucht}")

if "Berlin" in besucht:
    print("Berlin wurde schon besucht")

if "Hamburg" not in besucht:
    print("Hamburg wurde noch nicht besucht")
````

### Entscheidungstabelle: Welche Datenstruktur wann?

| Datenstruktur | Einsatz | Beispiel (Schiffe versenken) |
|---------------|---------|------------------------------|
| **Liste** | Geordnete Sammlung, Zugriff per Index | Spielfeldzeile, 2D-Spielfeld |
| **Tupel** | Festes Wertepaar, unveränderlich | Koordinate `(2, 3)` |
| **Dictionary** | Zugriff per Name/Bezeichnung | Schiffe: `{"Kreuzer": [...]}` |
| **Set** | Einzigartigkeit prüfen, Duplikate vermeiden | Geschossene Koordinaten |

````python
# Alle vier Datenstrukturen im Schiffe-versenken-Kontext

# Liste von Listen: Spielfeld
spielfeld = [["~"] * 5 for _ in range(5)]

# Dictionary: Schiffe mit Namen
schiffe = {
    "Kreuzer": [(0, 1), (0, 2), (0, 3)],
    "U-Boot": [(3, 0), (4, 0)]
}

# Set: Bereits geschossene Koordinaten
geschossen = set()

# Tupel: Eine einzelne Koordinate
schuss = (0, 2)

print(f"Spielfeld-Typ: {type(spielfeld)}")
print(f"Schiffe-Typ: {type(schiffe)}")
print(f"Geschossen-Typ: {type(geschossen)}")
print(f"Schuss-Typ: {type(schuss)}")
````

### Aufgabe 2: Treffer einem Schiff zuordnen

Schreibe Code, der für eine gegebene Koordinate (als Tupel) prüft, welchem Schiff sie gehört. Iteriere mit `for name, positionen in schiffe.items()` und nutze `in`.

### Aufgabe 3: Schiff versenken

Entferne bei einem Treffer die Koordinate aus der Positionsliste des Schiffs (`remove()`). Prüfe danach, ob die Positionsliste leer ist (`len() == 0`). Falls ja, gib `"Kreuzer versenkt!"` (bzw. den Schiffsnamen) aus.

Teste: Schieße nacheinander auf alle Positionen des U-Boots `(3, 0)` und `(4, 0)`.

<details>
<summary>💡 Tipp</summary>

```python
for name, positionen in schiffe.items():
    if schuss in positionen:
        positionen.remove(schuss)
        if len(positionen) == 0:
            print(f"{name} versenkt!")
        else:
            print(f"Treffer! {name} getroffen!")
```
</details>

### Aufgabe 4: Schüsse als Set

Ersetze die Schussliste aus VL 3 durch ein Set. Nutze `add()` zum Hinzufügen und `in` zum Prüfen auf Doppelschüsse.

Ablauf:
1. Erstelle ein leeres Set `geschossen`
2. Simuliere 3 Schüsse (z.B. `(0, 1)`, `(2, 3)`, `(0, 1)`)
3. Vor jedem Schuss: Prüfe ob die Koordinate schon im Set ist
4. Falls ja: `"Bereits geschossen!"`, falls nein: zum Set hinzufügen

<details>
<summary>💡 Tipp</summary>

