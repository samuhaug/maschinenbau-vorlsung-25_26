## Block 1: Funktionen definieren & aufrufen (45 min)

### Grundsyntax

Eine Funktion wird mit `def` definiert, nimmt Parameter entgegen und gibt mit `return` ein Ergebnis zurück:

```python
def funktionsname(parameter1, parameter2):
    # Code der Funktion
    return ergebnis
```

Der Aufruf erfolgt mit dem Namen und den Argumenten in Klammern: `funktionsname(wert1, wert2)`

````python
# Funktion definieren
def addiere(a, b):
    ergebnis = a + b
    return ergebnis

# Funktion aufrufen
summe = addiere(3, 7)
print(f"3 + 7 = {summe}")
````

````python
# Funktion ohne Rückgabewert (nur Ausgabe)
def begruesse(name):
    print(f"Hallo, {name}!")

begruesse("Anna")
begruesse("Max")
````

### Warum Funktionen?

1. **Wiederverwenden:** Code einmal schreiben, mehrfach aufrufen
2. **Struktur:** Programm in benannte Schritte zerlegen → lesbarer
3. **Testen:** Einzelne Funktionen isoliert prüfen

Statt einem langen Skript mit 100 Zeilen wird der Code in übersichtliche Bausteine zerlegt.

### Vorgehen beim Schreiben einer Funktion

Drei Schritte:

1. **Beschreiben:** Was tut die Funktion? (Ein Satz reicht)
2. **Schnittstelle festlegen:** Welche Parameter braucht sie? Was gibt sie zurück?
3. **Implementieren:** Den Code schreiben

**Aufteilregel:** Wenn man "und" braucht um die Funktion zu beschreiben → in zwei Funktionen aufteilen.

````python
# Beispiel: Funktion Schritt für Schritt entwickeln

# 1. Beschreibung: "Prüft ob eine Koordinate im 5×5-Spielfeld gültig ist"
# 2. Schnittstelle: Parameter: zeile (int), spalte (int) → Rückgabe: True/False
# 3. Implementierung:

def ist_gueltig(zeile, spalte):
    return 0 <= zeile <= 4 and 0 <= spalte <= 4

# Testen
print(ist_gueltig(0, 3))   # True — im Spielfeld
print(ist_gueltig(5, 2))   # False — Zeile 5 gibt es nicht
print(ist_gueltig(-1, 0))  # False — negative Zeile
````

````python
# Funktion mit mehreren Schritten
def beschreibe_zahl(zahl):
    if zahl > 0:
        return "positiv"
    elif zahl < 0:
        return "negativ"
    else:
        return "null"

# Rückgabewert in Variable speichern und weiterverwenden
ergebnis = beschreibe_zahl(42)
print(f"42 ist {ergebnis}")

# Oder direkt in print() verwenden
print(f"-7 ist {beschreibe_zahl(-7)}")
````

## Block 2: Scope & Zusammenspiel (45 min)

### Lokale vs. globale Variablen

Variablen innerhalb einer Funktion sind **lokal** — sie existieren nur innerhalb der Funktion. Variablen außerhalb sind **global**.

````python
# Lokale Variable: existiert nur innerhalb der Funktion
def berechne():
    ergebnis = 42  # lokale Variable
    print(f"In der Funktion: {ergebnis}")

berechne()

# Diese Zeile würde einen Fehler erzeugen:
# print(ergebnis)  # NameError: 'ergebnis' is not defined
````

````python
# Globale Variable: kann in der Funktion gelesen werden
spielname = "Schiffe versenken"

def zeige_titel():
    print(f"=== {spielname} ===")  # Liest die globale Variable

zeige_titel()
````

### Warum `global` vermeiden?

Statt globale Variablen zu verändern, sollte man:
- Daten als **Parameter** in die Funktion hineingeben
- Ergebnisse als **`return`** zurückgeben

So bleibt jede Funktion unabhängig und vorhersagbar.

````python
# Schlecht: Globale Variable verändern
zaehler = 0

def erhoehe_schlecht():
    global zaehler  # VERMEIDEN!
    zaehler = zaehler + 1

erhoehe_schlecht()
print(f"Zähler (schlecht): {zaehler}")
````

````python
# Gut: Wert als Parameter rein, Ergebnis als return raus
def erhoehe_gut(zaehler):
    return zaehler + 1

mein_zaehler = 0
mein_zaehler = erhoehe_gut(mein_zaehler)
print(f"Zähler (gut): {mein_zaehler}")
````

### Funktionen mit Datenstrukturen

Funktionen können Listen, Dictionaries und andere Datenstrukturen als Parameter empfangen und zurückgeben.

**Wichtig:** Wenn eine Funktion eine Liste oder ein Dictionary verändert (z.B. ein Element ändert), wirkt sich das auch außerhalb der Funktion aus — Listen und Dicts werden als Referenz übergeben.

````python
# Funktion die ein Spielfeld verändert
def markiere_feld(spielfeld, zeile, spalte, symbol):
    spielfeld[zeile][spalte] = symbol

# Spielfeld erstellen
spielfeld = [["~"] * 3 for _ in range(3)]
print(f"Vorher: {spielfeld[1]}")

# Feld markieren
markiere_feld(spielfeld, 1, 2, "X")
print(f"Nachher: {spielfeld[1]}")
````

### Funktionen rufen andere Funktionen auf

Ein Programm wird in kleine Schritte zerlegt. Jede Funktion löst ein Teilproblem. Die Funktionen rufen sich gegenseitig auf:

````python
# Kleine Funktionen die zusammenspielen
def ist_gueltig(zeile, spalte):
    return 0 <= zeile <= 4 and 0 <= spalte <= 4

def ist_treffer(schiffe, zeile, spalte):
    koordinate = (zeile, spalte)
    for name, positionen in schiffe.items():
        if koordinate in positionen:
            return name
    return None

# Hauptlogik nutzt beide Funktionen
schiffe = {"Kreuzer": [(0, 1), (0, 2), (0, 3)]}

zeile = 0
spalte = 2

if ist_gueltig(zeile, spalte):
    getroffenes_schiff = ist_treffer(schiffe, zeile, spalte)
    if getroffenes_schiff:
        print(f"Treffer! {getroffenes_schiff} getroffen!")
    else:
        print("Wasser!")
else:
    print("Ungültige Koordinate!")
````
