## 1. Variablen, `print`, Arithmetik (V01)

- Eine Variable ist ein benannter Speicherplatz: `drehzahl = 1500`.
- Ausgabe: `print(drehzahl)`.
- Die wichtigsten Rechenoperatoren: `+`, `-`, `*`, `/`, `**` (Potenz), `%` (Rest), `//` (ganzzahlige Division).

```python
drehzahl = 8000          # 1/min
durchmesser_mm = 80       # mm
umfang_mm = 3.14159 * durchmesser_mm
schnittgeschwindigkeit = umfang_mm * drehzahl / 1000  # m/min
print("Schnittgeschwindigkeit:", round(schnittgeschwindigkeit, 1), "m/min")
```

## 2. `input()` + Typumwandlung (V01)

- `input()` liefert **immer einen String** zurück – auch wenn du eine Zahl eingibst.
- Daher musst du mit `int(...)` oder `float(...)` umwandeln, bevor du rechnen kannst.
- Im Notebook simulieren wir die Eingabe als feste Variable (damit die Zelle automatisch durchläuft).

```python
# Statt input() setzen wir den Wert direkt – im echten Skript stünde:
#   eingabe = input("Drehzahl in 1/min: ")
eingabe = "2400"
drehzahl = int(eingabe)        # String -> int
print("Typ der Eingabe :", type(eingabe).__name__)
print("Typ nach int()  :", type(drehzahl).__name__)
print("Drehzahl + 100  :", drehzahl + 100)
```

```python
print(eingabe + eingabe)
print(drehzahl + drehzahl)

print("Typ der Eingabe :", type(eingabe))
```

## 3. f-Strings (V02)

Für saubere Ausgaben verwendest du **f-Strings**: vor den String ein `f`, Variablen in geschweiften
Klammern. Mit `:.2f` bestimmst du die Dezimalstellen.

```python
temperatur = 67.8429
druck_bar = 12.3
print(f"Temperatur: {temperatur:.1f} °C, Druck: {druck_bar:.2f} bar")
# Tipp: {x:>8.2f} -> rechtsbündig auf Breite 8, 2 Nachkommastellen
```

Formatiere `verschleiss_prozent` als f-String mit **einer** Nachkommastelle und dem Suffix `"%"`,

```python
verschleiss_prozent = 12.3456
# TODO: Ergebnis-String bauen
ausgabe = f"{verschleiss_prozent:.1f}%"  # <-- hier anpassen, z.B. f"{...:.1f}%"

try:
    assert ausgabe == "12.3%", f"Erwartet '12.3%', erhalten '{ausgabe}'"
    print("OK:", ausgabe)
except AssertionError as e:
    print("❌ Noch nicht richtig:", e)
```

## 4. Dateien lesen mit `with open(...)` (V02)

Mit dem `with`-Statement wird die Datei nach dem Block automatisch geschlossen –
auch wenn ein Fehler auftritt. Das ist die **empfohlene** Art, Dateien zu öffnen.

```python
# Wir lesen die Template-Datei aus dieser Vorlesung ein (existiert bereits im Ordner).
with open("cad_template_technisch.txt", encoding="utf-8") as f:
    inhalt = f.read()

print("Anzahl Zeichen:", len(inhalt))
print("Erste 120 Zeichen:")
print(inhalt[:120])
```

## 5. String-Methoden `.strip()` und `.split()` (V02)

- `.strip()` entfernt Leerzeichen, Tabs und Zeilenumbrüche am Anfang und Ende.
- `.split(sep)` zerlegt einen String an `sep` in eine Liste.

```python
zeile = "  Maschine-7 ;  42,5 ; OK  \n"
sauber = zeile.strip()
teile = [t.strip() for t in sauber.split(";")]
print("sauber :", repr(sauber))
print("teile  :", teile)
```
