## Block 1: `for`-Schleife (45 min)

### `range()` — Zahlenfolgen erzeugen

Die `for`-Schleife wiederholt Code für jedes Element einer Sequenz. Mit `range()` erzeugst du eine Zahlenfolge:

| Aufruf | Erzeugt | Beschreibung |
|--------|---------|-------------|
| `range(5)` | 0, 1, 2, 3, 4 | 5 Zahlen, startet bei 0 |
| `range(1, 6)` | 1, 2, 3, 4, 5 | Von 1 bis 5 (Endwert exklusive) |

**Merke:** Der Endwert ist immer **exklusive** — `range(1, 6)` liefert 1 bis 5, nicht bis 6.

````python
# range(n) — startet bei 0
for i in range(5):
    print(i)
````

````python
# range(start, stop) — Bereich festlegen
for i in range(1, 6):
    print(f"Zeile {i}")
````

### Iteration über Sequenzen

Eine `for`-Schleife kann auch direkt über die Zeichen eines Strings iterieren:

````python
# Über einen String iterieren
for buchstabe in "ABCDE":
    print(buchstabe)
````

### `enumerate()` — Index und Wert gleichzeitig

Manchmal brauchst du sowohl den **Index** (Position) als auch den **Wert**. Dafür gibt es `enumerate()`:

````python
# enumerate() liefert Index und Wert als Paar
spalten = "ABCDE"

for index, buchstabe in enumerate(spalten):
    print(f"Spalte {index}: {buchstabe}")
````

### Verschachtelte Schleifen

Für 2D-Ausgaben (z.B. ein Raster) braucht man eine **äußere Schleife** (Zeilen) und eine **innere Schleife** (Spalten):

```
Zeile 1: Spalte A, Spalte B, Spalte C, ...
Zeile 2: Spalte A, Spalte B, Spalte C, ...
...
```

Die innere Schleife läuft für **jede Iteration** der äußeren Schleife komplett durch.

````python
# Verschachtelte Schleife: 3×3 Raster
for zeile in range(3):
    for spalte in range(3):
        print("*", end=" ")  # end=" " verhindert den Zeilenumbruch
    print()  # Zeilenumbruch nach jeder Zeile
````

````python
# Spielfeld-Kontext: Spaltenköpfe + Raster
# Spaltenköpfe
print("  ", end="")
for buchstabe in "ABCDE":
    print(f" {buchstabe}", end="")
print()  # Zeilenumbruch

# Zeilen mit Zeilennummern
for zeile in range(1, 4):  # nur 3 Zeilen als Demo
    print(f"{zeile} ", end="")
    for spalte in range(5):
        print(" ~", end="")
    print()
````

### `print()` mit `end`-Parameter

Standardmäßig fügt `print()` am Ende einen Zeilenumbruch ein. Mit `end` kann man das ändern:

- `print("Hallo", end=" ")` → kein Zeilenumbruch, stattdessen ein Leerzeichen
- `print()` → nur ein Zeilenumbruch (nützlich nach einer Zeile im Raster)

## Block 2: `while`-Schleife (45 min)

### `for` vs. `while`

| | `for` | `while` |
|---|-------|--------|
| **Wann?** | Anzahl der Durchläufe ist **bekannt** | Abbruchbedingung entscheidet |
| **Beispiel** | "Gib 5 Zeilen aus" | "Frage bis gültige Eingabe" |
| **Syntax** | `for x in sequenz:` | `while bedingung:` |

````python
# while-Schleife: Zähler von 1 bis 5
zaehler = 1

while zaehler <= 5:
    print(f"Durchlauf {zaehler}")
    zaehler = zaehler + 1  # Wichtig: Zähler erhöhen!
````

### Endlosschleifen erkennen und vermeiden

Eine `while`-Schleife läuft **endlos**, wenn die Bedingung nie `False` wird:

```python
# ACHTUNG: Endlosschleife!
zaehler = 1
while zaehler <= 5:
    print(zaehler)
    # zaehler wird nie erhöht → Bedingung bleibt True
```

**Faustregel:** In jeder `while`-Schleife muss sich etwas ändern, das die Bedingung beeinflusst.

Falls eine Endlosschleife passiert: In Jupyter Notebook den **Kernel unterbrechen** (Stopp-Button ⬜).

### `break` — Schleife vorzeitig beenden

`break` beendet die Schleife sofort, egal ob die Bedingung noch `True` ist:

````python
# break: Abbruch wenn gewünschter Wert gefunden
for zahl in range(1, 100):
    if zahl * zahl > 50:
        print(f"{zahl} ist die erste Zahl deren Quadrat > 50 ist")
        break
````

### `continue` — Durchlauf überspringen

`continue` überspringt den Rest des aktuellen Durchlaufs und springt zum nächsten:

````python
# continue: Ungerade Zahlen überspringen
for zahl in range(1, 11):
    if zahl % 2 != 0:  # % ist der Modulo-Operator (Rest der Division)
        continue
    print(f"{zahl} ist gerade")
````

### Muster: Eingabeschleife mit `while True` + `break`

Ein häufiges Muster: Frage den Nutzer wiederholt, bis eine gültige Eingabe kommt.

```python
while True:
    eingabe = input("Eingabe: ")
    if eingabe_ist_gueltig:
        break
    print("Ungültige Eingabe, versuche es erneut.")
```

Das `while True` läuft endlos — der `break` in der `if`-Bedingung ist der einzige Ausgang.

````python
# Beispiel: Passwort-Eingabe (simuliert mit fester Prüfung)
# Zum Testen: gib "geheim" ein

while True:
    passwort = input("Passwort eingeben: ")
    if passwort == "geheim":
        print("Zugang gewährt!")
        break
    print("Falsches Passwort, nochmal.")
````
